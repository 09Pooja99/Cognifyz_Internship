# New Section

```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
```

```python
path = "/content/drive/MyDrive/Cognifyztechnologies_Internship/dataset .csv"
df= pd.read_csv(path)
```

```python
print(df.shape)
print(df.columns)
```

```python
print(df['Cuisines'])
```

```python
cuisines_count = df['Cuisines'].str.split(',').explode('Cuisines').value_counts()
print(cuisines_count)
```

```python
top_cuisines = cuisines_count.head(3)
print(top_cuisines)
```

```python
colors= ['skyblue', 'green', 'blue']
plt.bar(top_cuisines.index, top_cuisines.values, color=colors)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Top Three Cuisines')
plt.show()
```

Calculate the percentage of restaurants that serve each of the top cuisines.


```python
total_restaurants = len(df)
print('total restaurants are', total_restaurants)
top_cuisines = cuisines_count.head(3)
print('Top Three Cuisines are :', top_cuisines)
top_cuisines_percentage = (top_cuisines/ total_restaurants) * 100
print(' the percentage of restaurants that serve each of the top cuisines :', top_cuisines_percentage)
```

```python
color = ['red','orange', 'green']
plt.bar( top_cuisines.index, top_cuisines_percentage, color=color)
plt.xlabel('Cuisine')
plt.ylabel('Percentage')
plt.title('Restaurants that serve each of the top cuisines')
plt.show()
```
