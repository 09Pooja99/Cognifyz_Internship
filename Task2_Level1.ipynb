Identify the city with the highest number of restaurants in the dataset.


Calculate the average rating for restaurants in each city.

Determine the city with the highest average rating.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
path = ('/content/drive/MyDrive/Cognifyztechnologies_Internship/dataset .csv')
df = pd.read_csv(path)
```

```python
df.columns
```

```python
restaurant_counts = df['City'].value_counts()
print(restaurant_counts)
```

```python
restaurant_num = restaurant_counts.max()
city_with_highest_restaurants = restaurant_counts.head(1)
print('City with Highest restaurants: ', city_with_highest_restaurants)


```

Calculate the average rating for restaurants in each city.

```python
average_rating = df.groupby('City')['Aggregate rating'].mean()
print('Avrage rating for restaurants in each city : ', average_rating)
```

Determine the city with the highest average rating.

```python
highest_average_rating = average_rating.max()


print(highest_average_rating)


```
